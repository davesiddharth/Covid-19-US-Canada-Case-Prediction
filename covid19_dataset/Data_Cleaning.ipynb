```python
import pandas as pd
```

```python
csv_paths=[]
for i in range(22,32,1):
    csv_paths.append(f"01-{i}-2020.csv")

for i in range(1,10,1):
    csv_paths.append(f"02-0{i}-2020.csv")

for i in range(10,30,1):
    csv_paths.append(f"02-{i}-2020.csv")    

for i in range(1,10,1):
    csv_paths.append(f"03-0{i}-2020.csv")
    
for i in range(10,12,1):
    csv_paths.append(f"02-{i}-2020.csv")

```

```python
data_frame=[]
for path in csv_paths: 
    data_frame.append(pd.read_csv(path))
    
    
```

```python
result=pd.concat(data_frame)
```

```python
result.drop(columns=["Latitude","Longitude"],inplace=True)
result.head(500)

```

```python
result_copy=result.copy()
```

```python
result_copy=result_copy.fillna(0)
result_copy
```

```python
result_1=result_copy.groupby(["Country/Region","Last Update"])
```

```python
result_2=pd.DataFrame({"Cummulative Confirmed per Day":result_1["Confirmed"].sum(),\
                       "Cummulative Deaths per Day":result_1["Deaths"].sum(),"Cummulative Recovered cases per Day":result_1["Recovered"].sum()})
```
