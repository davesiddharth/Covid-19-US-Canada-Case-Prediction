```python
#Importing the dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Storing the csvpaths
csvpath_confirmed="time_series_19-covid-Confirmed.csv"
csvpath_death="time_series_19-covid-Deaths.csv"
csvpath_recovered="time_series_19-covid-Recovered.csv"
```

```python
#Reading into a pandas DataFrame
confirmed_df=pd.read_csv(csvpath_confirmed)
death_df=pd.read_csv(csvpath_death)
recovered_df=pd.read_csv(csvpath_recovered)
```

```python
#Manipulating the Confirmed dataframe
#Dropping the colums which are not required
confirmed_df.drop(columns=["Province/State","Lat","Long"],inplace=True)
```

```python
#Filling NaN with a 0 for calculations
confirmed_df=confirmed_df.fillna(0)
confirmed_df.head()
#len(Confirmed_df1
confirmed_df["Country/Region"]
```

```python
#Grouping the Confirmed cases dataframe by Country/Region
confirmed_group=confirmed_df.groupby(["Country/Region"])


#Creating the final Confirmed dataframe to be used further
confirmed_final=confirmed_group.sum()
confirmed_final.head()
#len(Confirmed_final)
```

```python
#Manipulating the Deaths dataframe
#Dropping the colums which are not required
death_df.drop(columns=["Province/State","Lat","Long"],inplace=True)
```

```python
#Filling NaN with a 0 for calculations
death_df=death_df.fillna(0)
death_df.head()
#len(Confirmed_df1)
```

```python
#Grouping the Death cases dataframe by Country/Region
death_group=death_df.groupby(["Country/Region"])


#Creating the final Death dataframe to be used further
death_final=death_group.sum()
death_final.head()
#len(Deaths_final)
```

```python
#Manipulating the Recovered dataframe
#Dropping the colums which are not required
recovered_df.drop(columns=["Province/State","Lat","Long"],inplace=True)
```

```python
#Filling NaN with a 0 for calculations
recovered_df=recovered_df.fillna(0)
recovered_df.head()
#len(Recovered_df1)
```

```python
#Grouping the Recovered cases dataframe by Country/Region
recovered_group=recovered_df.groupby(["Country/Region"])


#Creating the final Recovered dataframe to be used further
recovered_final=recovered_group.sum()
recovered_final.head(110)
#len(Recovered_final)
```

```python
len(recovered_final.loc["Korea, South"])
```

```python
#Creating a dataframe for South Korea
south_korea_df=pd.DataFrame({"Date":confirmed_final.loc["Korea, South",:].index.tolist(),\
                           "Cumulative Confirmed":confirmed_final.loc["Korea, South",:].tolist(),\
                           "Cumulative Death":death_final.loc["Korea, South",:].tolist(),
                           "Cumulative Recovered":recovered_final.loc["Korea, South",:].tolist()})
```

```python
south_korea_df.head()
```

```python
confirmed_total_south_korea=max(south_korea_df["Cumulative Confirmed"])
confirmed_total_south_korea
```

```python
death_total_south_korea = max(south_korea_df["Cumulative Death"])
death_total_south_korea
```

```python
recovered_total_south_korea=max(south_korea_df["Cumulative Recovered"])
recovered_total_south_korea
```

```python
total_population_south_korea = 51269185 
#total population in south korea

total_days = len(south_korea_df)
total_days
# total days since the start of the spread
```

```python
#% of Death=Total Death/Total confirmed *100
#% of Recovered=Total Recovered/Total confirmed *100
#Infection rate=Total confirmed/Total population *100
#Speed of Infection=Total Confirmed/Total days

infection_rate_south_korea = round((confirmed_total_south_korea/total_population_south_korea)*100,5)
#infection_rate_southkorea

percentage_death_south_korea = round((death_total_south_korea/confirmed_total_south_korea)*100,2)
#percentage_death_southkorea

percentage_recovered_south_korea = round((recovered_total_south_korea/confirmed_total_south_korea)*100,2)
#percentage_recovered_southkorea


speed_infection_south_korea = round((confirmed_total_south_korea/total_days),1)
speed_infection_south_korea
```

```python
plt.figure(figsize=(20, 5))
x_axis = np.arange(0, len(south_korea_df))
tick_locations = []
for x in x_axis:
    tick_locations.append(x)
    
plt.title("Cumulative Covid-19 cases in South Korea ")
plt.xlabel("Period in Days (From January 22, 2020)")
plt.ylabel("Number of Cases")
plt.grid()
    
confirmed_cases, = plt.plot(x_axis, south_korea_df.loc[:,'Cumulative Confirmed'], color="blue", label="Cumulative Confirmed Cases")
death_cases, = plt.plot(x_axis, south_korea_df.loc[:,'Cumulative Death'], color="red", label="Cumulative Death Cases")
recovered_cases, = plt.plot(x_axis, south_korea_df.loc[:,'Cumulative Recovered'], color="green", label="Cumulative Recovered Cases")

plt.legend(handles=[confirmed_cases, death_cases,recovered_cases], loc="upper left")
plt.xticks(tick_locations)
plt.xlim(0, 51)
plt.savefig("Cumulative Cases for South Korea.jpg")
plt.show()

print(f"{round(speed_infection_south_korea,0)} people per day got infected throughout {total_days} days.")
print(f"The rate at which the population got infected from 22nd Jan 2020 to 13th March 2020 was {infection_rate_south_korea}.")
print(f"{percentage_death_south_korea}% population suffered death from 22nd Jan 2020 to 13th March 2020.")
print(f"{percentage_recovered_south_korea}% population recovered from 22nd Jan 2020 to 13th March 2020.")
print(f"{infection_rate_south_korea}% people got infected out of the entire population of {total_population_south_korea}.")
```

```python
# Calculating the increase in confirmed cases per day 
south_korea_df["Date"]
cumulative_cases = south_korea_df["Cumulative Confirmed"]
difference = np.diff(cumulative_cases)
```

```python
difference.tolist()
south_korea_daily_df=pd.DataFrame({"Date":south_korea_df["Date"].drop(0).tolist(),\
                           "Daily Confirmed":difference.tolist()
                           })
south_korea_daily_df.head()
```

```python
# Plot the daily updates in South Korea

plt.figure(figsize=(20, 5))
x_axis = np.arange(0, len(south_korea_daily_df))
tick_locations = []
for x in x_axis:
    tick_locations.append(x)
    
plt.title("Daily Updates of Covid-19 Cases in South Korea ")
plt.xlabel("Period in Days (From January 23, 2020)")
plt.ylabel("Number of Cases")
plt.grid()
    
confirmed_cases, = plt.plot(x_axis, south_korea_daily_df.loc[:,'Daily Confirmed'], color="blue", label="Daily Confirmed Cases")

plt.legend(handles=[confirmed_cases], loc="upper left")
plt.xticks(tick_locations)
plt.xlim(0, 51)
plt.savefig("Daily Updated Cases for South Korea.jpg")
plt.show()
```
